```
df2 = df[["message"]]
df2 = df2.head(2000)
```

```
import re
import nltk
import itertools
import pandas as pd
from nltk import word_tokenize
from nltk.corpus import stopwords
from string import digits,punctuation 
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords') 
nltk.download('averaged_perceptron_tagger')

remove_digits = str.maketrans('', '', digits)
stop_words = set(stopwords.words('english'))

def tokens(str):
  str = str.lower()
  token = word_tokenize(str)
  words = [word for word in token if word not in stop_words and word not in punctuation]
  return words
```

```
df2['message1'] = df2['message'].apply(tokens)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
def dummy_fun(doc):
    return doc
vectorizer = TfidfVectorizer(analyzer='word',tokenizer=dummy_fun,preprocessor=dummy_fun,token_pattern=None)
vectors = vectorizer.fit_transform(df2['message1'])
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df3 = pd.DataFrame(denselist, columns=feature_names)
```

```
df3
```

```
from sklearn.decomposition import NMF
model = NMF(n_components=5, init='nndsvd', random_state=42, alpha = 0.1, l1_ratio = 0.5)
W = model.fit_transform(df3)
H = model.components_
```

```
print(W)
print(H)
```

```
import numpy as np
def display_topics(H, W, feature_names, documents, no_top_words, no_top_documents):
    for topic_idx, topic in enumerate(H):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        top_doc_indices = np.argsort( W[:,topic_idx] )[::-1][0:no_top_documents]
        for doc_index in top_doc_indices:
            print(documents[doc_index])
```
